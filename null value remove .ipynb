```python
import pandas as pd 
df=pd.read_csv("zomato_restaurants_in_India.csv")
df
```

```python
df1=df.drop(columns=["url","currency","zipcode"])
df1
```

```python
df2=df1.fillna(value=0)
df2
```

```python
df3=df2.drop_duplicates(subset='res_id')
df3
```

```python
df3.isnull()
```

```python
df3.isnull().sum()
```

```python
df4=df3.replace(to_replace=[],value=" ")
df4
```

```python
df4.isnull().sum()
```

```python
df4.to_csv("nullvaluepy.csv")
```
